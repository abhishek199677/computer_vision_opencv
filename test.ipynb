## hsv scale 

```python

import matplotlib.pyplot as plt
import cv2

hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

plt.imshow(hsv, cmap='hsv')   #hue-saturation-value(hsv)
plt.axis('off')                # Hide axes
plt.show()                     # Show the figure
```

```python

import matplotlib.pyplot as plt
import cv2

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # Convert to grayscale

plt.imshow(gray, cmap='gray')  # Display as grayscale
plt.axis('off')                # Hide axes
plt.show()                     # Show the figure
```

## rotate image

```python
(h,w) = img.shape[:2]
```

```python
h
```

```python
w
```

### rotated the image by 45 degress

```python
rotate = cv2.getRotationMatrix2D((w//2,h//2),45,1)

rotated = cv2.warpAffine(img, rotate, (w, h))

plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB))  # For color images
plt.axis('off')
plt.show()
```

### rotated the image by 180 degress

```python
rotate = cv2.getRotationMatrix2D((w//2,h//2),180,0.9)

rotated = cv2.warpAffine(img, rotate, (w, h))

plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB))  # For color images
plt.axis('off')
plt.show()
```

### rotated the image by 90 degrees and chnaged the image position from centre to 4

```python
rotate = cv2.getRotationMatrix2D((w//4,h//4),90,1)

rotated = cv2.warpAffine(img, rotate, (w, h))

plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB))  # For color images
plt.axis('off')
plt.show()
```

### blurred image

```python
blur = cv2.blur(img,(50,50))

blur_rgb = cv2.cvtColor(blur, cv2.COLOR_BGR2RGB)

plt.imshow(blur_rgb)
plt.axis('off')
plt.title('blur')
plt.show()
```

### gaussian blurring

```python

gaussian = cv2.GaussianBlur(img,(51,51),0)

# gaussian_rgb = cv2.cvtColor(gaussian, cv2.COLOR_BGR2RGB)
median = cv2.medianBlur(img,51)



plt.imshow(median)
plt.axis('off')
plt.show()
```
